```python
np.random.seed(20)
```

```python
x1= np.random.rand(1000,1)
x2= np.random.rand(1000,1)
x3= np.random.rand(1000,1)
x4= np.random.rand(1000,1)
x= np.concatenate((x1,x2,x3, x4), axis=1)
```

```python
t1, t2, t3, t4, b = 4, -2.5, 7, .1, -1
```

```python
def func(t1,x1,t2,x2,t3,x3, t4, x4, b):
    return b + t1*x1 + t2*x2 + t3*x3 + t4* x4
```

```python
y = func(t1,x1,t2,x2,t3,x3, t4, x4, b)
```

```python
#y
```

```python
model=tf.keras.Sequential()
model.add(keras.layers.Dense(units=1, input_shape=[4]))
```

```python
model.compile(optimizer='adam', loss= 'mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=500)
```

```python
sx1, sx2, sx3, sx4 = 0.5, 0.1, 0.7, 0.115 #example test values
```

```python
p = model.predict([[sx1, sx2, sx3, sx4]])
```

```python
real = func(t1, sx1, t2, sx2, t3, sx3, t4, sx4, b) # actual formula with test values
print("accuracy= ", float(p/real*100), '%')
```
